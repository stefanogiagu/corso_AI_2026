**Parte 1: Importazione del dataset in un dataframe pandas e analisi statistica e visuale del dataset**

Il dataset è costituito da 10M di eventi appartenenti a due processi fisici prodotti nelle interazioni protone-protone al colider LHC:

* produzione di jet da QCD: $q \to jet$ (**classe fondo**)
* produzione di due jet boosted da decadimento elettrodebole di un bosone $W$: $W \to qq' \to jet\,jet$ (**classe segnale**)

Un jet è un insieme collimato di particelle (mesoni, barioni, leptoni, fotoni), prodotto nella adronizzazione di quark e gluoni di alta energia prodotti nelle collisioni agli ecceleratori di particelle.

Nel caso dei decadimenti del $W$ a causa dell'alto boost di lorentz i due jet tendono a sovrapporsi in un singolo jet detto "merged" o "fat jet", rendendo gli eventi simili a quelli del processo di fondo. Tuttavia, la presenza dei due jet sovrapposti crea delle sotto-strutture peculiari all'interno del deposito di energia del jet ricostruito che possono essere utilizzate per distinguere i "merged" jet dai jet di QCD.

Ogni evento del dataset rappresenta un jet, descritto da 6 osservabili fisiche di alto livello (feature) e da una label $1,0$ che indica se l'evento è di classe segnale o di classe fondo.

Le osservabili di alto livello sono:

*   ***Mjet***: massa invariante del Jet
*   ***C21, C21, D21, D22***: 4 funzioni di correlazione energetica tra i sotto-elementi del jet
*   ***Tau21***: 21-subjetiness (una misura di quanto il jet è composto da 2 vs 1-subjet), tende a $1$ per jet composti da un solo jet e a 0 per jet composti da due jet.

I dati sono contenuti nel file *JetSubStruct.dat*, che è un file *.csv* in cui ogni riga contiene le 6 osservabili + la label di un evento.

**NOTA: Gli algoritmi come Naive Bayes e sopratutto k-NN possono diventare molto lenti al crescere della dimensione del campione di training, conviene quindi leggere solo una parte del dataset (i primi 10k eventi) per non rendere troppo lento l'esercizio**

````python
#@title Soluzione Parte 1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix

#importa il file .csv JetSubStruct.dat in un dataframe pandas
# nrows=10000 permette di leggere solo le prime 10k righe del file
df = pd.read_csv('JetSubStruct.dat', nrows=10000)

#creo una lista con le feature (escludendo la label)
colonne = list(df.columns)
features = colonne[:-1] #esclude l'ultimo elemento (la label)
print("colonne presenti nel dataframe:")
print(colonne)
print()

#printa il contenuto della tabella
print("Contenuto del dataframe")
display(df)
print()

#statistica descrittiva delle colonne della tabella
print("Tabella con la statistica descrittiva")
display(df.describe())
print()
#matrice di correlazione (escludendo la colonna delle label)
print("Tabella con la correlazione tra le osservabili")
display(df[features].corr())
print()


#istogrammi delle osservabili

# seleziono gli eventi della classe con label 1
signal = df.loc[df['Label']>0.5]
# seleziono gli eventi della classe con label 0
background = df.loc[df['Label']<0.5]

# lista di range opportuni per gli istogrammi delle varie osservabili
mrange = [[0,200],[0,0.5],[0,0.2],[0,5],[0,5],[0,1]]

print("Distribuzioni delle osservabili")
# questo comando permette di fissare le dimensioni della figura
plt.figure(figsize=(6,6))
for i in range(6):
  plt.subplot(3,2,i+1)
  plt.hist(signal[features[i]], bins=50, range=mrange[i], color='red', histtype='step', density=True, label='Signal')
  plt.xlabel(features[i])
  plt.hist(background[features[i]], bins=50, range=mrange[i], color='blue', histtype='step', density=True, label='Background')
  plt.xlabel(features[i])
  plt.legend()
plt.tight_layout()
plt.show()
print()

#scatter plots tra le osservabili

print("Scatter plot tra le osservabili")
_ = scatter_matrix(signal[features], color='red', alpha=0.2)
_ = scatter_matrix(background[features], color='blue', alpha=0.2)
````

NOTA IMPORTANTE: i parametri per la normalizzazione delle feature vanno ottenuti con il solo campione di training in modo da evitare contaminazione di informazione del campione di training nel campione di test (*information leaks nel test set*).

Il test set deve fare le veci di un campione di eventi distribuito come quello ustilizzato nel training (IID), ma mai osservato in precedenza, **da utilizzare unicamente per la misura delle prestazioni finali del modello di ML addestrato**, va quindi normalizzato rispetto ai parametri misurati dal solo training set.


````python
#code
````

````python
#@title Soluzione Parte 2

#divido in training e test sample

#size datasets
ntot = len(df.index)

#divisione in training e test usando lo slicing
train = df[:int(ntot*0.8)].copy()
test = df[int(ntot*0.8):].copy()

train = train.reset_index(drop=True)
test = test.reset_index(drop=True)

#normalizzo solo rispetto al training set
#normalizzazione ottenuta sottraendo la media e dividendo perla sdev (eg normalizzazione a N(0,1))
train_mean = train[features].mean()
train_std  = train[features].std()
train[features] = (train[features]-train_mean)/train_std
test[features] = (test[features]-train_mean)/train_std

#controlliamo che ora le feature siano normalizzate come atteso

print ('Media osservabili training set dopo normalizzazione')
print(train[features].mean())
print()
print ('Deviazione standard osservabili training set dopo normalizzazione')
print(train[features].std())
print()

print ('Media osservabili test set dopo normalizzazione')
print(test[features].mean())
print()
print ('Deviazione standard osservabili test set dopo normalizzazione')
print(test[features].std())
````

````python
#@title Soluzione 3A
#Naive Bays Classifier

from sklearn.naive_bayes import GaussianNB

model1 = GaussianNB()
model1.fit(train[features], train['Label'])

accuracy_train = model1.score(train[features], train['Label'])
accuracy_test = model1.score(test[features], test['Label'])

print('Naive Bays:')
print('Training accuracy: ', accuracy_train)
print('Test     accuracy: ', accuracy_test)
````

````python
#@title Soluzione 3B
#k-NN Classifier
from sklearn import neighbors

print('k-NN:')
print('k:  Training/Test accuracy: ')
acc = []
acc_test = []
kval = []
for k in np.linspace(1,30,10, dtype=int):
  model2 = neighbors.KNeighborsClassifier(k)
  model2.fit(train[features], train['Label'])

  accuracy_train = model2.score(train[features], train['Label'])
  accuracy_test = model2.score(test[features], test['Label'])
  acc.append(accuracy_train)
  acc_test.append(accuracy_test)
  kval.append(k)
  print(f'{k:2d}: {accuracy_train:4.2f}     {accuracy_test:4.2f}')
print('')
print('Massima accuracy sul test set ottenuta per k = ', kval[acc_test.index(max(acc_test))],'\n')

plt.plot(kval,acc, 'r.', label='training set')
plt.plot(kval,acc_test, 'b.', label='test_set')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.legend()
plt.show()
````

````python
#@title Soluzione 3C
#LogisticRegression

from sklearn.linear_model import LogisticRegression

# penalty='l2' regolarisation L2
# C: Inverse of regularization strength (smaller values means stronger regularisation)
# solver: algorithm to use in the optimization problem (lbfgs: second order gradient method)

print('Logistic Regression:')

logreg = LogisticRegression(penalty='l2', C=0.01, solver='lbfgs', max_iter=100000)
logreg.fit(train[features], train['Label'])

accuracy_train = logreg.score(train[features], train['Label'])
accuracy_test = logreg.score(test[features], test['Label'])

print('Training accuracy: ', accuracy_train)
print('Test     accuracy: ', accuracy_test)
````

**Parte 4: Costruzione e allenamento di un modello di regressione di tipo k-NN**

L'algoritmo k-NN può essere utilizzato per predire un target continuo (regressione). In questo caso funziona stimando il target di una funzione y = f(x), come media dei k valori del campione di training più vicini (in accordo ad una predefinita distanza $d(x,x_i)$) al valore $x$:
> $\hat{y} = \hat{f}(x) = \frac{1}{k} \sum_{i\in k-nn(x)} y_i$. <p>

Compiti:

1.   addestrare un modello scikit-learn di tipo n*eighbors.KNeighborsRegressor* per la task di predire il valore di Mjet a parire dalle altre 5 features (tau,C,D). Testare tre valori del parametro $k$:  $k=1,5,10, 20, 100$, e per oguno di essi printare il valore del coefficiente di determinazione $R^2$:

>```
># coefficiente di determinazione:
>```
>$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$:  
> ```
> in cui:
> ```
> $SS_{tot}$ = residui riespetto al valore medio del traget nei dati $(y-<y>) ^2$ (questa è la predizione più naive che si può pensare di fare) <p>
> $SS_{res}$ = residui rispetto alla predizione del modello $(y-pred)^2$ <p>
> ```
> assume valore 0 se il modello performa come un modello naive, 1 se il modello performa meglio del naive, e assume valori negativi se performa peggio del modello naive
>
>#ottenibile tramite il metodo .score(...) applicato al  modello KNeighborsRegressor
>```


NOTA: esempio di uso come regressore di k-NN disponibile nel sito di scikit-learn: [link](https://scikit-learn.org/stable/auto_examples/neighbors/plot_regression.html#sphx-glr-auto-examples-neighbors-plot-regression-py)


2.   graficare i valori predetti per Mjet VS i valori veri nel training e nel test set, per il modello che fornisce il migliore coefficient di determinazione.

````python
#code
````

````python
#@title Soluzione 4
#k-NN Regressor
from sklearn import neighbors

variabili = features[1:] #le variabili sono tutte le feasture tranne la prima feature (Mjet)
target = features[0]     #il target è la prima feature (Mjet)

plt.figure(figsize=(10,10))

best_R2 = -999.0
best_k = 1
pred_train_best = []
pred_test_best = []
print('  k R2 score')
for index,k in enumerate([1,5,10,20,100]):
  model3 = neighbors.KNeighborsRegressor(k)
  model3.fit(train[variabili].values, train[target].values)

  # coefficient of determination
  # R^2 = 1 - SSres/SStot:  0 if a model perform as the naive model, 1 for the best case, negative for worst than naive
  # SStot = residuals wrt data average (y-<y>)^2
  # SSres = residuals wrt model prediction (y-pred)^2
  R2 = model3.score(test[variabili].values,test[target].values)
  print(f'{k:3d} {R2:.3f}')

  pred_train = model3.predict(train[variabili].values)
  pred_test =  model3.predict(test[variabili].values)

  if R2 >= best_R2:
    best_R2 = R2
    best_k = k
    pred_train_best = pred_train
    pred_test_best = pred_test

# plot per il best model
print("Max R^2 score for k = ", best_k)
print(" ")
plt.figure(figsize=(6,4))

plt.subplot(1,2,1)
plt.title('Training set k='+str(best_k))
plt.scatter(train[target].values,pred_train_best, s=0.1)
plt.plot([-2,5],[-2,5], color='black', linewidth=1, linestyle='dashed')
plt.ylabel('Mjet predetta')
plt.xlabel('Mjet vera')

plt.subplot(1,2,2)
plt.title('Test set k='+str(best_k))
plt.scatter(test[target].values, pred_test_best, s=0.1)
plt.plot([-2,5],[-2,5], color='black', linewidth=1, linestyle='dashed')
plt.ylabel('Mjet predetta')
plt.xlabel('Mjet vera')
plt.tight_layout()
plt.show()
````
