## Introduction: the 2D Ising model

The **Ising model** is one of the standard models used in statistical physics to study collective behavior, phase transitions, and Monte Carlo methods. In its two-dimensional version, the system is defined on a square lattice of size $L \times L$. Each site $i$ carries a **spin** variable

$$
\sigma_i \in {-1, +1},
$$

which can be interpreted as a microscopic magnetic moment pointing either **down** ($-1$) or **up** ($+1$).

The energy of a spin configuration is described by the Hamiltonian

$$
H({\sigma}) = -J \sum_{\langle i,j \rangle} \sigma_i \sigma_j,
$$

where the sum runs over **nearest-neighbor pairs** and $J>0$ corresponds to the ferromagnetic case. In this case, aligned neighboring spins lower the energy, so the system tends to form ordered domains.

At finite temperature, thermal fluctuations compete with this energetic tendency toward alignment. As a result, the model exhibits two qualitatively different regimes:

* at **low temperature**, spins tend to align and the system develops a nonzero magnetization;
* at **high temperature**, thermal disorder dominates and the average magnetization tends to zero.

A useful observable is the **magnetization**

$$
M = \frac{1}{N} \sum_i \sigma_i,
$$

with $N = L^2$, which measures the global degree of alignment of the lattice.

In these exercises, the goal is to simulate the Ising model using the **Metropolis algorithm**, a Markov Chain Monte Carlo method. Starting from an initial spin configuration, we repeatedly propose single-spin flips and accept or reject them according to the corresponding energy change $\Delta E$ and the Boltzmann weight. This allows us to generate configurations distributed according to thermal equilibrium.

This setting is useful for understanding several core ideas at once:

* how microscopic update rules generate macroscopic behavior;
* how the Metropolis acceptance rule depends on temperature;
* how observables such as magnetization and energy evolve during the simulation;
* how the system behaves near the phase transition.
